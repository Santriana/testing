## Area Pots (Series/Dataframe) <a id="12"></a>

**What is a line plot and why use it?**

An Area chart or area plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. Use line plot when you have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

**Questions:**
1. Bagaimana tren kejian kasus kriminalitas di Top 3 Wilayah pada tahun 2010


```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
filtered_df = df[df['year'] == 2010]

# Hitung total kriminalitas per borough dan ambil borough dengan total tertinggi
top_boroughs = filtered_df.groupby('borough')['value'].sum().sort_values(ascending=False).head(3).index

# Filter DataFrame lagi untuk top 3 borough
filtered_df_top3 = filtered_df[filtered_df['borough'].isin(top_boroughs)]

# Visualisasi menggunakan plot area
filtered_df_top3.pivot_table(values='value', index='month', columns='borough', aggfunc='sum').plot(kind='area',
                                                                                                alpha=0.25,
                                                                                                stacked=False,
                                                                                                figsize=(20, 10))

# Graded-Funtion End
plt.title('Tren kriminalitas di Top 3 Wilayah pada tahun 2010')
plt.ylabel('Jumlah Kriminalitas')
plt.xlabel('Bulan')
plt.show()
```

**Insight:**

Pada grafik dapat dilihat bahwa tren peningkatan kriminalitas untuk 3 top wilayah di London di setiap bulannya pada tahun 2010 mengalami peningkatan yang fluktuatif. Pada wilayah Westminster peningkatan terjadi cukup signifikan hal ini dikarenakan jumlah populasi di wilayah Westminster yang terbesar diantara kedua wilayah lainnya yaitu sebesar 89.701 jiwa. Sedangkan untuk wialayah Lambeth dan Camden peningkatan kasus kriminalitas terjadi cukup stagnan hal ini dikarenakan jumlah populasi di wilayah tersebut bekiras sebanyak 49.645 jiwa

## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

**Question:**
1. Bagaimana persebaran tahun terjadinya major category "Burglary"

```python
# Write your function below

# Graded-Funtion Begin (~2 Lines)
burglary_2010_df = df[(df['major_category'] == 'Burglary') & (df['year'] == 2010)]

# Visualisasi menggunakan histogram
plt.figure(figsize=(12, 8))
sns.histplot(burglary_2010_df['value'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.axvline(burglary_2010_df['value'].mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(burglary_2010_df['value'].median(), color='green', linestyle='dashed', linewidth=2, label='Median')

# Graded-Funtion End
plt.title('Distribusi Nilai Kejahatan "Burglary" pada Tahun 2010')
plt.xlabel('Jumlah Kejahatan')
plt.ylabel('Frekuensi')
plt.legend()
plt.show()
```

**Insight:**
Berdasarkan grafik histogram tersebut dapat diketahui bahwa nilai kejahatan "burglary" pada tahun 2010 terkonsentrasi pada value 0 hal ini ditunjukkan dengan nilai median yang berada pada value 0. Hal ini berarti pada tahun 2010 jumlah kasus untuk kejahatan "burglary" jarang terjadi dikarenakan persebaran data terkonsentrasi pada value 0 dengan variabel value merepresentasikan jumlah kejahatan untuk wilayah tertentu, pada bulan tertentu.

## Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**
1. Bagaimana perbandingan jumlah data untuk masing - masih major category pada tahun 2010

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
df_2010 = df[(df['year'] == 2010)]

# Group by minor_category dan sum value
grouped_data = df_2010.groupby('major_category')['value'].sum()
plt.bar(grouped_data.index, grouped_data)

# Graded-Funtion End
plt.xlabel('Major Category')
plt.ylabel('Jumlah')
plt.title('Jumlah Major Category pada tahun 2010')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Insight:**
Berdasarkan hasil barchart tersebut dapat dilihat bahwa jumlah setiap major category pada tahun 2010 terdapat perbedaan yang sangat signifikan. Pada grafik ditunjukkan tidak terjadi kasus untuk kategori sexual offences dan Fraud or forgery. Sedangkan pada tahun tersebut jumlah kasus Theft and Handling sangat tinggi yang disusul dengan kategori violence against the person.

## Pie Charts <a id="6"></a>

A `pie chart` is a circualr graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices. You are most likely already familiar with pie charts as it is widely used in business and media. We can create pie charts in Matplotlib by passing in the `kind=pie` keyword.

**Question:**
 

```python
# Write your function below
df_major = df[(df['borough'] == 'Westminster')].groupby('major_category', axis=0).sum()

# ratio for each continent with which to offset each wedge.
explode_list = [0.1] * len(df_major)

# Graded-Funtion Begin (~8 Lines)
df_major['value'].plot(kind='pie',
                       figsize=(15, 6),
                       autopct='%1.1f%%',
                       startangle=90,
                       shadow=True,
                       labels=None,
                       pctdistance=1.12,
                       explode=explode_list
                       )
# Graded-Funtion End

# scale the title up by 12% to match pctdistance
plt.title('Persebaran Major Categories di Westminster', y=1.12)

plt.axis('equal')

# add legend
plt.legend(labels=df_major.index, loc='upper left')
plt.show()
```

**Insight:**
Berdasarkan piechart dapat disimpulkan bahwa pada Boroguh Westminster kategori Burglary memiliki persentase jumlah kasus terbesar dengan persentase sebesar 61%. Sedangkan disisi lain pada Boroguh Westminster tidak tercatat kasus Fraud and Forgery dan terdapat sedikit kasus sexual offences dengan persentase 0.1%

## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

**Question:**
1. Kasus apa yang tergolong outlier pada major category pada tahun 2010?

```python
df_box = df[(df['year'] == 2010)].groupby('major_category')['value'].sum()

df_box
```

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
df_box.plot(kind='box',
                figsize=(8,6))

# Graded-Funtion End

plt.title('Boxplot Major Category Pada Tahun 2010')
plt.ylabel('Jumlah Kasus')

plt.show()
```

**Insight:**
Pada boxplot dapat dilihat bahwa kasus Theft and Handling tergolong sebagai outlier. Hal ini dikarenakan jumlah kasus tersebut secara signifikan berbeda jauh dibandingkan dengan kasus - kasus lainnya. Sedangkan untuk jumlah kasus minimum yaitu kasus Fraud or Forgery dan Sexual Offences. Kemudian untuk jumlah kasus maksimum yaitu kasus Violence Against Person. Selanjutnya untuk kasus tergolong median yaitu kasus Criminal Damage.

## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

**Question:**
1. Bagaimana korelasi antara peningkatan kasus dengan peningkatan tahun

```python
df_scatter = df[df['borough'] == 'Westminster'].groupby(df['year'])['value'].sum().reset_index()
```

```python
df_scatter['populasi'] = df_scatter['value'] * 2
```

```python
df_scatter
```

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
df_scatter.plot(kind='scatter', x='populasi', y='value', figsize=(10, 6), color='darkblue')
# Graded-Funtion End

plt.title('Korelasi Antara Peningkatan Kasus Dengan Peningkatan Tahun')
plt.xlabel('Populasi')
plt.ylabel('Jumlah Kasus')

plt.show()
```

**Insight:**
Pada grafik scatter dapat disimpulkan bahwa terdapat korelasi yang kuat antara populasi dengan jumlah kasus dimana peningkatan pada setiap tahunnya untuk jumlah kasus kriminalitas dikarenakan peningkatan jumlah populasi juga.

## Word Clouds <a id="8"></a>


`Word` clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

```python
# install wordcloud
# !conda install -c conda-forge wordcloud --yes

!pip install wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
stopwords = set(STOPWORDS)
```

```python
# instantiate a word cloud object
text = " ".join(minor_category for minor_category in df['minor_category'])

your_wordcloud = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
your_wordcloud.generate(text)
```

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
plt.imshow(your_wordcloud, interpolation='bilinear')
# Graded-Funtion End

plt.axis('off')
plt.show()
```

**Insight:**
Pada wordcloud data dilihat bahwa terdapat beberapa kasus dengan frekuensi yang tinggi. Hal ini ditunjukkan dengan jenis kasus dengan font kata yang besar seperti Criminal Damage, Theft Taking, Motor Vehicle, dan Common Assault
